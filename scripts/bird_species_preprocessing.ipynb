# Importing Libraries

```python
import pandas as pd
```

# Importing The Forest Dataset and Combining All the Sheets

```python
forest_sheets = pd.read_excel('/content/Bird_Monitoring_Data_FOREST.XLSX', sheet_name=None)

# Combining into one DataFrame
forest_df = pd.concat(forest_sheets.values(), ignore_index=True)
print(forest_df.shape)
```

# Importing The Grassland Dataset and Combining All the Sheets

```python
grassland_sheets = pd.read_excel('/content/Bird_Monitoring_Data_GRASSLAND.XLSX', sheet_name=None)

# Combining into one DataFrame
grassland_df = pd.concat(grassland_sheets.values(), ignore_index=True)
print(grassland_df.shape)
```

# Standardizing Columns

```python
forest_df.columns = forest_df.columns.str.strip().str.lower()
grassland_df.columns = grassland_df.columns.str.strip().str.lower()
```

# Adding Source Column

```python
forest_df['habitat'] = 'forest'
grassland_df['habitat'] = 'grassland'
```

# Merging The Datasets

```python
combined_df = pd.concat([forest_df, grassland_df], ignore_index=True)
```

# Cleaning and Optimizing the Dataset

```python
# Removing duplicates
df = combined_df.drop_duplicates().reset_index(drop=True)


# Handling missing values
df.loc[:, 'sex'] = df['sex'].fillna('Unknown')
df.loc[:, 'flyover_observed'] = df['flyover_observed'].fillna('False')
```

# Creating Time Features

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.columns = df.columns.str.strip().str.lower()
```

# Exporting Datasets

```python
df.to_csv("bird_cleaned_data.csv", index=False)
```
